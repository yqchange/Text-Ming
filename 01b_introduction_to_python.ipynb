# Exceptions

```python
def to_int (x):
    return int(x)
```

```python
to_int("5")
```

```python
to_int("5.5")
```

```python
def to_int_safe (x):
    try:
        z = int (x)
    except:
        print ("We cannot convert everything like this!")
        z = -1
    return z
```

```python
to_int_safe("5.5")
```
